```python
# Group data by country and select the maximum values for each country (total cases and total deaths)
df_top = df.groupby('location')[['total_cases', 'total_deaths']].max()

# Select the top 10 countries with the highest number of cases and deaths
top_cases = df_top.nlargest(10, 'total_cases')  # Top 10 by total cases
top_deaths = df_top.nlargest(10, 'total_deaths')  # Top 10 by total deaths

# Convert values to millions for better readability
top_cases['total_cases'] /= 1e6
top_deaths['total_deaths'] /= 1e6

# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot total cases (left subplot)
top_cases['total_cases'].plot(kind='barh', ax=axes[0], color='blue', edgecolor='black')
axes[0].set_title("Top 10 Countries by Total COVID-19 Cases")  # Chart title
axes[0].set_xlabel("Total Cases (Millions)")  # X-axis label
axes[0].invert_yaxis()  # Invert Y-axis to show country with highest cases at top

# Plot total deaths (right subplot)
top_deaths['total_deaths'].plot(kind='barh', ax=axes[1], color='red', edgecolor='black')
axes[1].set_title("Top 10 Countries by Total COVID-19 Deaths")  # Chart title
axes[1].set_xlabel("Total Deaths (Millions)")  # X-axis label
axes[1].invert_yaxis() 

# Format x-axis labels with one decimal place
for ax in axes:
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.1f}'))

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
data2 = df.copy()
data2.date = pd.to_datetime(data2['date'])
data2 = data2.groupby('date').sum()
data2['7 days MA new cases'] = 0
data2['7 days MA new cases'] = data2['new_cases'].rolling(7).mean() #Moving average of new cases with window=10
data2['7 days MA new deaths'] = 0
data2['7 days MA new deaths'] = data2['new_deaths'].rolling(7).mean()  #Moving average of new deaths with window=10

data2[['new_cases', '7 days MA new cases']].plot(figsize = (11, 5), alpha = 0.5)
plt.title('Timeline new cases in world')
plt.xlabel('Date')
plt.ylabel('New cases')
```
